```python
tweets = df.text
tweets
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
```

```python
from nltk import TweetTokenizer
tk = TweetTokenizer()
tweets = tweets.apply(lambda x:tk.tokenize(x)).apply(lambda x:" ".join(x))
tweets
```

```python
#regular expression eg:

str1 = 'Wonderfull @peacock@#123!!'
str2 = re.sub('[^a-zA-Z0-9]'," ",str1)
str2 
```

```python
# to Remove Special Character

tweets = tweets.str.replace('[^a-zA-Z0-9]+',' ')         # ('[^a-zA-Z0-9]+',' ') it is a reguler expression pattern.
tweets                                                          # '+' to skip the combination of special character.
```

```python
from nltk.tokenize import word_tokenize
tweets = tweets.apply(lambda x:' '.join([w for w in word_tokenize(x) if len(w)>=3]))
tweets
```

```python
from nltk.stem import SnowballStemmer
ss = SnowballStemmer('english')
tweets = tweets.apply(lambda x:[ss.stem(i.lower()) for i in tk.tokenize(x)]).apply(lambda x:' '.join(x))
tweets
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
tweets = tweets.apply(lambda x:[i for i in tk.tokenize(x) if i not in stop]).apply(lambda x:' '.join(x))
tweets
```

```python
#vectorization :

from sklearn.feature_extraction.text import TfidfVectorizer
vec = TfidfVectorizer()
train_data = vec.fit_transform(tweets)

print(train_data)
```

```python
train_data.shape
```

```python
y = df['target'].values
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(train_data,y,test_size=.30,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

knn_model = KNeighborsClassifier()
svm_model = SVC()
bayes_model = MultinomialNB()
tree_model = DecisionTreeClassifier()
rb_model = RandomForestClassifier()
lsb_model = [knn_model,svm_model,bayes_model,tree_model,rb_model]
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
for i in lsb_model:
  print(i)
  i.fit(xtrain,ytrain)
  y_pred = i.predict(xtest)
  print('*'*10)
  print(accuracy_score(ytest,y_pred))
```
